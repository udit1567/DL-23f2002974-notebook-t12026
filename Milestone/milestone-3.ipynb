```python
import torch
import torch.nn as nn
import torchaudio
import numpy as np
```

### You have a batch of grayscale spectrogram images represented as a PyTorch tensor with shape (32, 1, 64, 64). You need to flatten this tensor (excluding the batch dimension) to feed it into a Fully Connected (Linear) layer. What is the size of the input features for that Linear layer?

```python
spectrograms = torch.randn(32, 1, 64, 64)

# Flatten each spectrogram while keeping the batch dimension
# -1 tells PyTorch to automatically calculate the remaining size
flattened_spectrograms = spectrograms.view(32, -1)

# Print shapes
print("Original shape:", spectrograms.shape)
print("Flattened shape:", flattened_spectrograms.shape)

# Number of features that will go into a Linear layer
input_features = flattened_spectrograms.shape[1]
print("Input features for Linear layer:", input_features)
```

### You are converting a 1-second audio clip (Sample Rate = 16,000 Hz) into a Mel-Spectrogram using torchaudio. The configuration is n_fft=400, hop_length=160, n_mels=64. Based on your code output, what is the resulting tensor shape?



```python
# Create 1 second of random audio
sample_rate = 16000
audio = torch.randn(1, sample_rate)   # (channels, samples)

# Define Mel Spectrogram transformation
mel_spec = torchaudio.transforms.MelSpectrogram(
    sample_rate=sample_rate,
    n_fft=400,
    hop_length=160,
    n_mels=64
)

# Convert audio → Mel Spectrogram
mel = mel_spec(audio)

# Print shapes
print("Audio shape:", audio.shape)
print("Mel spectrogram shape:", mel.shape)
```

### Your CNN passes a (1, 64, 64) input through nn.Conv2d(..., kernel_size=3, stride=1, padding=1) followed by nn.MaxPool2d(kernel_size=2, stride=2). What is the spatial output shape (Height, Width)?

```python
# Create a dummy input image
# Shape: (batch_size, channels, height, width)
x = torch.randn(1, 1, 64, 64)

# Convolution layer
# padding=1 keeps the spatial size the same
conv_layer = nn.Conv2d(in_channels=1, out_channels=16, kernel_size=3, stride=1, padding=1)

# Max pooling layer
# This reduces height and width by half
pool_layer = nn.MaxPool2d(kernel_size=2, stride=2)

# Apply convolution first
after_conv = conv_layer(x)

# Then apply pooling
output = pool_layer(after_conv)

# Print shapes
print("Input shape:", x.shape)
print("After Conv:", after_conv.shape)
print("After MaxPool:", output.shape)

# Spatial size (height, width)
print("Spatial output (H, W):", output.shape[2], output.shape[3])
```

### You have a hidden layer defined as nn.Linear(in_features=128, out_features=64). How many total learnable parameters (weights + biases) does this specific layer contain?

```python
# Create a Linear layer
linear_layer = nn.Linear(128, 64)

# Count parameters
num_weights = linear_layer.weight.numel()   # total weights
num_biases = linear_layer.bias.numel()      # total biases

total_params = num_weights + num_biases

# Print results
print("Weights:", num_weights)
print("Biases:", num_biases)
print("Total parameters:", total_params)
```

### Your training dataset consists of 1,050 audio files. You initialize a DataLoader with batch_size=32 and drop_last=False. How many batches will run in one epoch?

```python
import math

# dataset size and batch size
total_files = 1050
batch_size = 32

# number of batches needed
num_batches = math.ceil(total_files / batch_size)

print("Total files:", total_files)
print("Batch size:", batch_size)
print("Number of batches:", num_batches)
```

### You are performing a 3-class classification. The model outputs logits [2.5, 1.0, 0.1] for a target of Class 0. Using nn.CrossEntropyLoss, what is the calculated loss value (rounded to 3 decimals)?


```python
# Model output (logits for 3 classes)
logits = torch.tensor([[2.5, 1.0, 0.1]])

# Correct class label
target = torch.tensor([0])

# Cross entropy loss
loss_fn = nn.CrossEntropyLoss()

# Compute loss
loss = loss_fn(logits, target)

print("Logits:", logits)
print("Target class:", target.item())
print("Loss:", round(loss.item(), 3))
```

### You are training with SGD. Current Weight = 0.5, Gradient = 0.2, Learning Rate = 0.01. What is the new weight value after one optimizer step?

```python
# values from the model
weight = 0.5
gradient = 0.2
lr = 0.01   # learning rate

# SGD update
new_weight = weight - lr * gradient

print("Current weight:", weight)
print("Gradient:", gradient)
print("Learning rate:", lr)
print("Updated weight:", new_weight)
```

### You want to apply a convolution with kernel_size=5 and stride=1 such that the output size remains the same as the input. What padding value must you set?



```python
# kernel size of convolution
kernel_size = 5

# padding needed to keep same spatial size
padding = (kernel_size - 1) // 2

# create dummy input image
x = torch.randn(1, 1, 64, 64)

# convolution layer
conv = nn.Conv2d(1, 1, kernel_size=kernel_size, stride=1, padding=padding)

# run convolution
output = conv(x)

print("Kernel size:", kernel_size)
print("Padding:", padding)
print("Input shape:", x.shape)
print("Output shape:", output.shape)
```
